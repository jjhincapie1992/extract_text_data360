```python
import json
import time
from os import listdir
from requests import get, post
```

```python
def ls(ruta = '.'):
    return listdir(ruta)

carpeta = 'C:/Users/Juan Jose Hincapie/Google Drive/Transel Clasificacion/Centro/aceites/formato_2019/'
path = ls(carpeta)
path
files = []
for i in range(0,len(path)):
    if ('.pdf' in path[i]) == True:
        files.append(path[i])
    else:
        pass

files

#como leerlos
#informe = json_to_table("C:/Users/user/Desktop/Data360/Prueba Texto Chile/DataNueva/"+files[i])
```

```python
########### Python Form Recognizer Async Analyze #############

# Endpoint URL
endpoint = r""
apim_key = ""
model_id = ""
post_url = endpoint + "/formrecognizer/v2.0/custom/models/%s/analyze" % model_id
source = r"direccion/" + files[0]
params = {
    "includeTextDetails": True
}
#33
headers = {
    # Request headers
    'Content-Type': 'application/pdf',
    'Ocp-Apim-Subscription-Key': apim_key,
}
with open(source, "rb") as f:
    data_bytes = f.read()

try:
    resp = post(url = post_url, data = data_bytes, headers = headers, params = params)
    if resp.status_code != 202:
        print("POST analyze failed:\n%s" % json.dumps(resp.json()))
        quit()
    print("POST analyze succeeded:\n%s" % resp.headers)
    get_url = resp.headers["operation-location"]
except Exception as e:
    print("POST analyze failed:\n%s" % str(e))
    quit()
```
